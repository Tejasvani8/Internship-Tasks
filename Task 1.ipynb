The task is to predict the percentage of marks that a student is expected to score based upon the number of hours they studied.


```python
# Importing the required libraries
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Step 1: Reading the data 


```python
# Reading data from remote link
url = r"https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
s_data = pd.read_csv(url)
print("Data import successful")

s_data.head(10)
```

# Step 2: Input data visuals 


```python
# Plotting the distribution of scores
s_data.plot(x='Hours', y='Scores', style='s')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

From the graph we can assume a positive linear relation between the number of hours studied and percentage of score.

# Step 3: Data preprocessing

This step involves division of data into "attributes" (inputs) and "labels" (outputs).


```python
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values
```


# Step 4: Model Training

Splitting the data into training and testing sets, and training the algorithm.


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
regressor = LinearRegression()  
regressor.fit(X_train.reshape(-1,1), y_train) 

print("Training complete.")
```


# Step 5: Plotting the Line of regression

Now since the model is trained,it's time to visualize the best-fit line of regression.


```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line,color='black');
plt.show()
```


# Step 6: Making Predictions

Now that the algorithm is trained,it's time to test the model by making some predictions using the test data Set.


```python
# Testing data
print(X_test)
# Model Prediction 
y_pred = regressor.predict(X_test)
```

# Step 7: Comparing Actual result to the Predicted Model result

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df
```

```python
#Estimating training and test score
print("Training Score:",regressor.score(X_train,y_train))
print("Test Score:",regressor.score(X_test,y_test))
```

```python
# Plotting the Bar graph to depict the difference between the actual and predicted value

df.plot(kind='bar',figsize=(7,7))
plt.grid(which='major', linewidth='0.5', color='black')
plt.grid(which='minor', linewidth='0.5', color='red')
plt.show()
```

```python
# Testing the model with our own data
# What will be predicted score if a student studies for 9.25 hrs/ day?
hours = 9.25
test = np.array([hours])
test = test.reshape(-1, 1)
own_pred = regressor.predict(test)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```


# Step 8: Evaluating the model

The final step is to estimate the performance of algorithm. Here different errors have been calculated to compare the model performance and predict the accuracy.


```python
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-2:', metrics.r2_score(y_test, y_pred))
```

R-2 gives the score of model fit and in this case we have R-2 = 0.9454906892105355 which is fair score for this model.
